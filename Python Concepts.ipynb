Runnable In Python, there is no built-in Runnable interface. Instead, Python uses a structural design philosophy called "Duck Typing" (if it walks like a duck and quacks like a duck, it’s a duck). If an object has the mechanisms to be executed, Python treats it as a runnable entity. 
Callables (The Ultimate "Runnable" Concept)
In Python, the broadest definition of a runnable object is a Callable. A callable is any object that you can put parentheses () after to execute code.

Python determines if something is callable by checking if it implements a special hidden method called __call__.

Functions and Methods: These are natively callable.

Custom Classes as Runnables: You can turn any standard object into a runnable entity by defining the __call__ method inside it. This is highly useful for creating objects that need to maintain state across multiple executions.

```python
#Runnable in python 
class BackgroundWorker:

# What self indicates is equivalent to this statement in java , it gives identifier to the method class 
# when it goes for execution it will be able to identify the method , it is like thread id  
    def __init__(self, task_name):
        self.task_name = task_name
        self.execution_count = 0

    # This magic method makes the class instance "Runnable"
    # *args - This allows the method to accept any number of positional arguments
    # **kwargs - This allows the method to accept any number of keyword arguments
    def __call__(self, *args, **kwargs):
        self.execution_count += 1
        print(f"Executing {self.task_name}! (Run count: {self.execution_count})")

# Create the runnable object
worker = BackgroundWorker("Database Cleanup")
worker2= BackgroundWorker("Send emails")

# Run it just like a function
worker()  # Output: Executing Database Cleanup! (Run count: 1)
worker2()  # Output: Executing Database Cleanup! (Run count: 2)
```

An asynchronous operation is a task that can take some time to complete, but instead of making the program sit idle and wait, Python can work on other tasks while that operation is in progress. await functions is similar to asyncio , it resumes execution of main function unitl called function complteit wors

```python
# Before asysn tasks get executed sequentially 
import time

def task1():
    time.sleep(3)

def task2():
    time.sleep(3)

task1()
task2()

import asyncio

async def task1_async():
    print("task 1 statred")
    await asyncio.sleep(8)
    print ("after waiting 8 seconds")

async def task2_async():
    print("task2 started ")
    await asyncio.sleep(2)
    print ("after waiting 2 seconds")

async def main():
    await asyncio.gather(task1_async(), task2_async())

import nest_asyncio
nest_asyncio.apply()
asyncio.run(main())
```
